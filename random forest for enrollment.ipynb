```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import export_graphviz
import pydot
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
```

```python
data = pd.read_csv('~/Documents/fall2019/5753/enrollment_data.csv')
data.head()
```

```python
data.drop(['Unnamed: 0','Unnamed: 0.1','sex'], axis = 1)
```

```python
for i in range(0, len(data)):
    data.loc[i,'Target_Enroll'] = int(data.loc[i,'Target_Enroll'])
# change target variable to np array
labels = np.array(data['Target_Enroll'])
```

```python
X = data.loc[:, data.columns != 'Target_Enroll']
y = data.loc[:, data.columns == 'Target_Enroll']
```

```python
# save feature name for future use
feature_list = list(X.columns)
```

```python
X_features = np.array(X)
```

# Training and Testing Sets

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_features, y, test_size = 0.25,
                                                                           random_state = 42)
```

# did not clear max depth and criterion

```python
# Instantiate model 
rf = RandomForestRegressor(n_estimators= 1000, random_state=42)

# Train the model on training data
rf.fit(X_train,  y_train.values.ravel());
```

```python
y_pred = rf.predict(X_test)
```

```python
y_test = np.array(y_test)
```

```python
# Calculate the absolute errors
errors = abs(y_pred - y_test)

# Print out the mean absolute error (mae)
print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees.')
```

```python
# Import tools needed for visualization
from sklearn.tree import export_graphviz
import pydot

# Pull out one tree from the forest
tree = rf.estimators_[5]

# Export the image to a dot file
export_graphviz(tree, out_file = 'enroll_tree.dot', feature_names = feature_list, rounded = True, precision = 1)

# Use dot file to create a graph
(graph, ) = pydot.graph_from_dot_file('enroll_tree.dot')

# Write graph to a png file
graph.write_png('enroll_tree.png'); 
```
