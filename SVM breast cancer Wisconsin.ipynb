```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
from sklearn import datasets 
cancer = datasets.load_breast_cancer()
```

```python
df_cancer = pd.DataFrame(np.c_[cancer['data'], cancer['target']], columns = np.append(cancer['feature_names'],['target']))
df_cancer.head()
```

```python
sns.pairplot(df_cancer, hue = 'target', vars = ['mean radius', 'mean texture','mean perimeter','mean area','mean smoothness'])
```

```python
plt.figure(figsize = (20,12))
sns.heatmap(df_cancer.corr(),annot = True)
```

```python
X = df_cancer.drop(['target'], axis = 1)
X.head()
```

```python
y=df_cancer['target']
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=20)
```

```python
print('Thie size of our training "X" (input features) is', X_train.shape)
print('\n')
print('Thie size of our testing "X" (input features) is', X_test.shape)
print('\n')
print('Thie size of our training "y" (input features) is', y_train.shape)
print('\n')
print('Thie size of our testing "y" (input features) is', y_test.shape)
print('\n')
```

```python
from sklearn.svm import SVC
svc_model = SVC()
```

```python
svc_model.fit(X_train, y_train)
```

```python
y_predict = svc_model.predict(X_test)
```

# confusion matrix is TN, TP, FP, FN

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
cm = np.array(confusion_matrix(y_test, y_predict, labels = [1,0]))
confusion = pd.DataFrame(cm, index=['is_cancer','is_healthy'],
                        columns=['predicted_cancer','predicted_healthy'])
confusion
```

```python
sns.heatmap(confusion, annot = True)
```

```python
print(classification_report(y_test, y_predict))
```

```python
X_train_min = X_train.min()
X_train_min
```

```python
X_train_max = X_train.max()
X_train_max
```

```python
X_train_range = (X_train_max - X_train_min)
X_train_range
```

```python
X_train_scaled = (X_train - X_train_min)/(X_train_range)
X_train_scaled.head()
```

```python
X_test_min = X_test.min()
X_test_range = (X_test - X_test_min).max()
X_test_scaled = (X_test - X_test_min)/X_test_range
```

```python
svc_model = SVC()
svc_model.fit(X_train_scaled, y_train)
```

```python
y_predict = svc_model.predict(X_test_scaled)
cm = confusion_matrix(y_test,y_predict)
```

```python
cm = np.array(confusion_matrix(y_test, y_predict, labels=[1,0]))
confusion = pd.DataFrame(cm, index=['is_cancer','is_healthy'],
                        columns=['predicted_cancer','predicted_healthy'])
confusion
```
